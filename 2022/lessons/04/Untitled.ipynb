```python
import numpy as np
import pandas as pd

import iqplot

import colorcet

import bokeh.io
import bokeh.models
bokeh.io.output_notebook()
```

```python
df = pd.read_csv('../data/bee_sperm.csv', comment='#')
df.head()
```

```python
p = iqplot.ecdf(
    df,
    q='Alive Sperm Millions',
    cats='Treatment',
    conf_int=True,
    n_bs_reps=100_000,
)

bokeh.io.show(p)
```

```python
df.groupby('Treatment')['Alive Sperm Millions'].mean().reset_index()
```

```python
ctrl = df.loc[df['Treatment']=='Control', 'Alive Sperm Millions'].values
pest = df.loc[df['Treatment']=='Pesticide', 'Alive Sperm Millions'].values
```

```python
def draw_bs_sample(data):
    return np.random.choice(data, size=len(data))
```

```python
import scipy.special
```

```python
def draw_bs_reps(data, stat_fun, size=1, args=()):
    return np.array([stat_fun(draw_bs_sample(data), *args) for _ in range(size)])
```

```python
bs_reps_ctrl = draw_bs_reps(ctrl, np.median, size=100_000)
bs_reps_pest = draw_bs_reps(pest, np.median, size=100_000)
```

```python
np.percentile(bs_reps_ctrl, [2.5, 97.5])
```

```python
np.percentile(bs_reps_pest, [2.5, 97.5])
```

```python
np.percentile(bs_reps_ctrl - bs_reps_pest, [0.5, 99.5])
```
